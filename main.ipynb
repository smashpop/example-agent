```python
import openai
import os

movie_api_url = os.getenv('MOVIE_API_URL')

client = openai.OpenAI()

messages = []
```

```python
def call_ai():
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages = messages,
    )
    messaage = response.choices[0].message.content
    messages.append({"role": "assistant", "content": messaage})
    print(f"AI: {messaage}")
```

```python
while True:
    message = input("You: ??? ")
    if message.lower() in ["q", "quit"]:
        break
    else:
        messages.append({"role": "user", "content": message})
        print(f"You: {message}")
        call_ai()
```
