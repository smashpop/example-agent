```python
import openai
import os

movie_api_url = os.getenv('MOVIE_API_URL')

client = openai.OpenAI()

PROMPT = """
You are a movie information agent.
You can user following api url for more inforamtion : https://nomad-movies.nomadcoders.workers.dev
Just answer the question using the functions below. Just answer the functions to use, do not add your own information.

You know three functions:

get_popular_movies() → calls https://nomad-movies.nomadcoders.workers.dev/movies to fetch popular movies.

get_movie_details(id) → calls https://nomad-movies.nomadcoders.workers.dev/movies/id to fetch details of a movie.

get_movie_credits(id) → calls https://nomad-movies.nomadcoders.workers.dev/movies/id/credits to fetch cast and crew.

When the user asks “What are the current popular movies?”, call get_popular_movies() and answer with the movie titles. If the user asks for more details or credits, call the corresponding function and respond naturally with the information.

Answer the following question: 
ID 550 영화의 출연진이 누구야?
"""

respose = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": PROMPT}  
    ],
)

respose
```
